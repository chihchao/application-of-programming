### 載入資料集

```python
import pandas as pd

df = pd.read_csv('客戶聯絡狀況資料檔.csv')
df.head()
```

## Label Encoding

```python
df['學歷'].unique()
education = {'博士':6, '小學':0, '大學':5, '國中':1, '高中':3, '專科':4, '高職':2}
df['學歷'] = df['學歷'].map(education)
df.head()
```

```python
df['訂購'].unique()
buy = {'偶爾':2, '頻繁':3, '一次':1, '從未':0}
df['訂購'] = df['訂購'].map(buy)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# 選擇要進行 label encoding 的欄位
columns_to_encode = ['工作']

# 建立 LabelEncoder 物件
label_encoder = LabelEncoder()

# 針對每一欄要進行 label encoding 的欄位進行 encoding
for column in columns_to_encode:
    df[column] = label_encoder.fit_transform(df[column])

df.head()
```

## One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder

# 選擇要進行 One-hot encoding 的欄位
columns_to_encode = ['婚姻','連絡方式']

# 建立 OneHotEncoder 物件
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False, drop='first')
```

```python
# 使用 fit_transform 直接對多個欄位進行編碼
encoded_data = encoder.fit_transform(df[columns_to_encode])
encoded_data
```

```python
# 取得編碼後的欄位名稱
encoded_columns = encoder.get_feature_names_out(columns_to_encode)
encoded_columns
```

```python
# 建立新的 DataFrame 來儲存編碼後的資料
encoded_df = pd.DataFrame(encoded_data, columns=encoded_columns)
```

```python
# 將編碼後的資料與原始資料合併
df = pd.concat([df, encoded_df], axis=1)
df.head()
```

```python
# 移除原始的欄位
df = df.drop(columns=columns_to_encode)
df.head()
```
