### 載入鳶尾花資料集

```python
from sklearn.datasets import load_iris

iris = load_iris()
x = iris.data         # 特徵資料
y = iris.target       # 標籤（目標）
```

```python
iris.target_names
```

```python
import pandas as pd

# 建立一個DataFrame，包含特徵資料和標籤
df = pd.DataFrame(data=x, columns=iris.feature_names)
df['target'] = y

# 顯示前幾筆資料
print(df.head())
```

### EDA

```python
df.info()
```

```python
df['target'].unique()
```

### 分割訓練集與測試集

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### Standardization

```python
# knn 最好進行數值標準化
```

### 建立與訓練 KNN 模型

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)  # 可以調整 n_neighbors 參數
knn.fit(x_train, y_train)
```

### 預測與評估模型

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = knn.predict(x_test)
print("準確率：", accuracy_score(y_test, y_pred))
print("分類報告：\n", classification_report(y_test, y_pred))
```

預測

```python
# new_data = pd.DataFrame([[5.5,2.6,4.4,1.2]], columns=iris.feature_names)
import numpy as np
new_data = np.array([[5.5,2.6,4.4,1.2]])
# 預測
prediction = knn.predict(new_data)
predicted_label = iris.target_names[prediction[0]]

print("預測結果：", predicted_label)
```

### 模型的儲存與載入

```python
import joblib

# 儲存模型到檔案
joblib.dump(knn, 'knn_iris_model.pkl')
```

```python
# 從檔案載入模型
loaded_model = joblib.load('knn_iris_model.pkl')

# 用來預測
pred = loaded_model.predict(new_data)
print("預測結果：", iris.target_names[pred[0]])
```
