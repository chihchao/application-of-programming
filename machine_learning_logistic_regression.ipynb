## 癌症辨識

```python
import pandas as pd
df = pd.read_csv('/content/breastCancer.csv')
df
```

```python
x = df.drop(['種類'], axis=1)
y = df['種類']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
transfer = StandardScaler()
x_train = transfer.fit_transform(x_train)
x_test = transfer.transform(x_test)
logis = LogisticRegression()
logis.fit(x_train, y_train)
score = logis.score(x_test, y_test)
score
```

```python
pred = logis.predict(x_test)
pred
```

```python
from sklearn.metrics import classification_report
ret = classification_report(y_test, pred, labels=(1, 2),target_names=('良性', '惡性'))
print(ret)
```

改善召回力：改變判定的機率，提高判為惡性的機率

```python
# 列出所有判定的機率，並留下惡性的機率
pred_p = logis.predict_proba(x_test)
pred_p = pred_p[:, 1]
pred_p
```

```python
import numpy as np
pred = np.where(pred_p > 0.05, 2, 1)
ret = classification_report(y_test, pred, labels=(1, 2),target_names=('良性', '惡性'))
print(ret)
```
