```python
import pandas as pd

data = {
    'ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen', 'Ivy', 'Jack'],
    'Age': [25, 30, 35, 40, 22, 29, 38, 27, 33, 31],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose'],
    'Salary': [50000, 60000, 75000, 80000, 45000, 62000, 72000, 58000, 69000, 71000],
    'Department': ['IT', 'HR', 'Finance', 'Marketing', 'IT', 'HR', 'Finance', 'Marketing', 'IT', 'Finance']
}

df = pd.DataFrame(data)

df
```

## 資料檢索

### 取得 Name 和 Salary 二欄資料

```python
df[['Name','Salary']]
```

### 取出索引值 5 到 8 共 4 筆資料

```python
df.loc[5:8]
```

### 找出年齡和薪水都高於平均值的人員

```python
condition = (df['Age'] > df['Age'].mean()) & (df['Salary'] > df['Salary'].mean())
df[condition]
```

### 請問公司人員的平均薪水、及最高、最低薪水？

```python
print(df['Salary'].mean(), df['Salary'].max(), df['Salary'].min(), df['Salary'].median())
```

### 公司共有幾個部門？那個部門薪水最高？那個部門最年輕？

```python
df['Department'].unique()
```

```python
print(df.groupby('Department')['Salary'].mean().idxmax(), df.groupby('Department')['Salary'].mean().max())
```

```python
print(df.groupby('Department')['Age'].mean().idxmin(), df.groupby('Department')['Age'].mean().min())
```

### 年齡是否與薪資有關？

```python
corr = df[['Age','Salary']].corr()
corr
```

## 資料清理

### 將薪水高於第3四分位數的人員標示為高薪

```python
hs = df['Salary'].quantile(0.75)
hs
```

```python
df['HighSalary'] = df['Salary'].apply(lambda x: True if x > hs else False)
df
```

### 刪除高薪欄位

```python
df = df.drop('HighSalary', axis=1)
df
```

### 使用 pd.concat() 新增資料（資料集）

```python
new_data = [
    {'ID': 111, 'Name': 'Kevin', 'Age': 28, 'City': 'Austin', 'Salary': None, 'Department': 'IT'},
    {'ID': 112, 'Name': 'Lily', 'Age': 26, 'City': 'Seattle', 'Salary': 55000, 'Department': 'Marketing'},
    {'ID': 112, 'Name': 'Lily', 'Age': 26, 'City': 'Seattle', 'Salary': 55000, 'Department': 'Marketing'}
]
df_new = pd.DataFrame(new_data)
df_new
```

```python
df = pd.concat([df, df_new], ignore_index=True)
df
```

### 找出空值的資料列，並填充空值（薪水依該部門平均值）

```python
df_avg = df.groupby('Department')['Salary'].mean()
df_avg
```

```python
for i in range(len(df)):
    if pd.isna(df.loc[i, 'Salary']):
        df.loc[i, 'Salary'] = df_avg[df.loc[i, 'Department']]
df
```

### 刪除重複資料

```python
df =  df.drop_duplicates()
df
```

### 新增欄位 'NewSalary'，如果薪水小於平均值則薪水加 500，否則維持原薪水；新增欄位 'RankSalary' 填入薪水排名。

```python
df['NewSalary'] = df['Salary'].apply(lambda x: x + 500 if x < df['Salary'].mean() else x)
df
```

```python
df['RankSalary'] = df['Salary'].rank(ascending=False)
df
```

```python
df[['Salary','NewSalary','RankSalary']] = df[['Salary','NewSalary','RankSalary']].astype(int)
df
```
