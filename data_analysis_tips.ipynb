```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips.dtypes
```

```python
corr = tips[['total_bill', 'tip', 'size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True)
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='smoker')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='smoker', style='time')
plt.show()
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm')
plt.show()
```

```python
sns.catplot(data=tips, x='day', y='tip', hue='sex', kind='swarm')
plt.show()
```

```python
sns.scatterplot(data=tips, x='day', y='tip', hue='sex')
plt.show()
```

```python
sns.boxplot(data=tips, x='sex', y='tip', hue='sex')
plt.show()
```
